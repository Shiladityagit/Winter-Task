## Popularity based recommendation system

### (Top books of the platform)

```python
books_with_ratings=Ratings.merge(Books,on='ISBN')
```

```python
num_rating = books_with_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating
```

```python
books_with_ratings['Book-Rating'] = pd.to_numeric(books_with_ratings['Book-Rating'], errors='coerce')

```

```python

# Check data type
print(books_with_ratings['Book-Rating'].dtype)

# Convert data type to numeric
books_with_ratings['Book-Rating'] = pd.to_numeric(books_with_ratings['Book-Rating'], errors='coerce')

# Check for missing values
print(books_with_ratings['Book-Rating'].isnull().sum())

# Check for non-numeric values
print(books_with_ratings['Book-Rating'].unique())

# Group by 'Book-Title' and compute mean
avg_rating_df = books_with_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the column
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
```

```python
avg_rating_df
```

```python
popular_books_df = num_rating.merge(avg_rating_df,on='Book-Title')
popular_books_df
```

```python
popular_books_df = popular_books_df[popular_books_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_books_df = popular_books_df.merge(Books,on='Book-Title').drop_duplicates('Book-Title')
```

## Collaborative filtering based system

```python
books_with_ratings.groupby('User-ID').count()['Book-Rating']
```

```python
x=books_with_ratings.groupby('User-ID').count()['Book-Rating']>150
```

```python
good_users=x[x].index
```

```python
good_users
```

```python
filtered_rating = books_with_ratings[books_with_ratings['User-ID'].isin(good_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>50
famous_books=y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
 pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        print(pt.index[i[0]])
    
    return data
```

## Working of the model

```python
recommend('The Da Vinci Code')
```

```python
recommend('The Notebook')
```
